반복 -> 반복 대상 -> 값이 연속적으로 있다 -> Sequence

연속된 값에 번호를 부여할 수 있는 자료들 : Sequential Data(순차 데이터)

이때, 부여되는 번호를 인덱스 번호

인덱싱 : 인덱스 번호를 가지고 값에 접근하는 것을 얘기함, 대괄호 [ ] 사용

파이썬에서, 데이터 접근할 때 [ ]

기능을 사용할 때 ( )

슬라이싱 : 순차 데이터의 부분집합을 얻어 내는 것

[start:stop:step] : start ~ (stop-1)

start 기본값 : 0

stop 기본값 : 인덱스 마지막 번호+1 , -1

step 기본값 : 1

len() : Sequential Data의 길이를 반환해주는 함수

- 좋아하는 색을 리스트로 설정하고 이를 for로 반복하여 출력하라.
- 단어를 3개 이상 리스트로 설정하고 이를 ‘단어, 단어길이’를 출력하라(단어 길이는 len( )을 이용합니다).
- 리스트에 1,2,3,4,5를 저장하고, for 반복으로 2,3,4,5을 출력하여라.

```python
fav_color = ['파랑', '노랑', '검정']

for i in fav_color:
    print(i)

words = ['코딩', '파이썬', '데이터분석']

for i in words:
    print(i, len(i))

list2 = [1, 2, 3, 4, 5]

for i in list2[1:]:
    print(i)
```

• for와 range( )를 사용하여 0부터 10까지 출력하시오.

• for와 range( )를 사용하여 5부터 15까지 출력하시오.

• for와 range( )를 사용하여 0~21까지의 짝수를 모두 출력하시오.

• for와 range( )를 사용하여 –100~100까지의 4의 배수를 모두 출력하시오.

```python
for i in range(0,11):
    print(i)

for i in range(5,16):
    print(i)

for i in range(0, 22):
    if i%2 == 0:
        print(i)

for i in range(-100, 101):
    if i%4 == 0:
        print(i)
```
