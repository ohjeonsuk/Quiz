```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data 생성
data = np.arange(1, 11)

# 그래프 그리기
plt.plot(data)

# 그래프를 보여주는 코드 (주피터 노트북에서는 안해줘도 그래프 출력됨)
plt.show()
```

```python
data1 = np.arange(1, 11)
plt.plot(data1)

data2 = np.arange(11, 21)
plt.plot(data2)

plt.show()
```

```python
data1 = np.arange(1, 11)
plt.plot(data1)

# 캔버스  추가
plt.figure()

data2 = np.arange(11, 21)
plt.plot(data2)

plt.show()
```

```python
# 2행 2열로 canvas를 나누어 1번(1행 1열) 칸에 subplot 그리기
data1 = np.arange(1, 11)
plt.subplot(2, 2, 1)
plt.plot(data1)

# 2행 2열로 canvas를 나누어 3번(2행 1열) 칸에 subplot 그리기
data2 = np.arange(11, 21)
plt.subplot(2, 2, 3)
plt.plot(data2 * 3)

# 2행 2열로 canvas를 나누어 4번(2행 2열) 칸에 subplot 그리기
data3 = np.arange(21, 31)
plt.subplot(2, 2, 4)
plt.plot(data3 ** 2)

plt.show()
```
