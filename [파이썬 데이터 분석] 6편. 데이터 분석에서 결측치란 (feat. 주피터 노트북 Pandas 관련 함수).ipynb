```python
null_count = df.isnull().sum()
null_count
```

```python
# .reset_index() --> series를 data frame으로 변환
df_null_count = null_count.reset_index()
df_null_count.head()
```

```python
df_null_count.columns = ["컬럼명", "결측지수"]
df_null_count.head()
```

```python
#df에서 결측치가 가장 많은 10개를 df형식으로 가져옴.
df_null_count_top10 = df_null_count.sort_values(by="결측지수", ascending=False).head(10)
df_null_count_top10
```

```python
drop_columns = df_null_count_top10["컬럼명"]
drop_columns
```

```python
drop_columns_list = drop_columns.tolist()
drop_columns_list
```

```python
print(df.shape)
df_drop = df.drop(drop_columns, axis=1)
print(df_drop.shape)
```
